### Sentiment Analysis - Fake News Classifier
- By using LSTM Neural Network - predict whether the News is Fake or Not

```python
#Lets import the libraries -
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Lets load the dataset -
df = pd.read_csv('C:/Users/om/Desktop/train.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
#Lets Drop Nan Values because we cannot fix/handele NaN Values with text data and we have many records (20800)-
df=df.dropna()
```

```python
df.isnull().sum()
```

```python
#Lets split the data in x & y Axis veriables -
x=df.drop('label',axis=1)
y=df['label']
```

```python
#Lets set Vocabulary size randomly- Lets consider in our dictionary we have 5000 different words - you can take any -
voc_size=5000
```

```python
#Onehot Representation, it will give you index number of words -
messages=x.copy()
messages
```

```python
#Lets reset the index -
messages.reset_index(inplace=True)
```

```python
#Lets call by indexing -
messages['title'][0]
print(len(messages['title']))
```

```python
#Lets load the nltk library -
import nltk
nltk.download('punkit')
nltk.download('stopwords')
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
from nltk.corpus import stopwords
stopwords.words('english')
```

```python
#Lets Create corpus -
corpus = []
for i in range(0,18285):
    a1=nltk.word_tokenize(messages['title'][i])
    a1=' '.join(a1)
    corpus.append(a1)
corpus
```

```python
#Lets do the Text preprocessing -
import re
review = []
for i in range(len(corpus)):
    a2=re.sub('[^ a-zA-Z]',' ',corpus[i]).lower().split()
    a3=[lemmatizer.lemmatize(i) for i in a2 if not i in stopwords.words('english')]
    a3=' '.join(a3)
    review.append(a3)
review
```

```python
#Lets take a look on changes we made -
print(len(review))
print(messages['title'][0])
print(review[0])
```

```python
#Lets do Word Embedding i.e- One Hot Encoding to convert Words into Vectors-
from tensorflow.keras.preprocessing.text import one_hot
onehot_repr=[one_hot(words,voc_size)for words in review] 
onehot_repr
```

```python
#Lets take a look on the changes i.e- words vs vectors-
print(messages['title'][0])
print('===================')
print(onehot_repr[0])
```

```python
#Embedding Representation : Lets apply Padding (Pre/Post) to set/standardize sentence length -
#Lets apply Post-Padding - i.e- standardize the length of sentence and add zeros in short-length sentences after the sentence ends- 
from tensorflow.keras.preprocessing.sequence import pad_sequences
sent_length=20
embedded_docs=pad_sequences(onehot_repr,padding='post',maxlen=sent_length)
print(embedded_docs)
```

```python
print(embedded_docs[1])
print(embedded_docs[0])
```

```python
#Lets create an object for the algorithm -
import tensorflow as tf
from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
model=Sequential()
```

```python
#Lets Add an Layers and Features -
embedding_vector_features=40   #features representation - 40 features
model.add(Embedding(voc_size,embedding_vector_features,input_length=sent_length))
model.add(LSTM(100))           #LSTM Neural network layer with 100 input Neurons
model.add(Dense(1,activation='sigmoid'))   #for Binary output (Activation function = sigmoid, loss = binary_crossentropy)
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```

```python
#Lets see length -
len(embedded_docs),y.shape
```

```python
#Lets convert List to array -
import numpy as np
X_final=np.array(embedded_docs)
y_final=np.array(y)
print(X_final.shape,y_final.shape)
```

```python
#Lets split the data in train_test_split parts -
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size=0.33, random_state=42)
```

```python
# Lets Train the model - you can also apply early stopping -
model.fit(X_train,y_train,validation_data=(X_test,y_test),epochs=10,batch_size=64)
```

```python
#Lets Predict the Y - axis values -
y_pred=model.predict(X_test)
```

```python
#Lets consider if output value is less than 0.5 consider 0 and Else consider 1 - Because this is Binary Classification -
y_pred=np.where(y_pred > 0.5, 1,0)
```

```python
#Lets Evaluate our Model -
from sklearn.metrics import confusion_matrix,accuracy_score
print(confusion_matrix(y_test,y_pred))
print('=============================================')
print('accuracy_score:',accuracy_score(y_test,y_pred))
print('==============================================')
```
